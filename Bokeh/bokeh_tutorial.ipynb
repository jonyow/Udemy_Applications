```python
from bokeh.plotting import figure, output_file, show

import pandas as pd

df = pd.DataFrame(columns = ["X", "Y"])

df["X"] = list(range(5))
df["Y"] = [5,2,3,4,1]

p = figure() # (df, x="X", y='Y', title = 'Temp by Day', xlab="Day", ylab='Temperature')

p.circle(df["X"], df["Y"], size=[ i + 5 for i in df["X"]], color='red')

#output_file("Scatter.html")

show(p)
```

```python
import pandas as pd

df = pd.read_excel( "verlegenhuken.xlsx", sheetname=0)

df["Temperature"] = df["Temperature"] / 10
df["Pressure"] = df["Pressure"] / 10

df.head()
```

```python
p = figure()

p.circle(df["Temperature"], df["Pressure"], size=1, color='blue')

p.title.text= "Temp and Air Pressure"
p.xaxis.axis_label = "Temp (C)"
p.xaxis.minor_tick_line_color = None

p.yaxis.axis_label = "Pressure (hPa)"
p.yaxis.minor_tick_line_color = None


#output_file("Scatter.html")

show(p)
```

```python
#
import datetime

df["Datetime"] = df[['Year', 'Month', 'Day', 'Hour']].apply(lambda s : datetime.datetime(*s),axis = 1)
```

```python

p = figure(x_axis_type = "datetime")

p.line(df["Datetime"], df["Temperature"], color = 'Orange', alpha = 0.5)

p.line(df["Datetime"], df["Pressure"]/10 - 100 , color = 'Blue', alpha = 0.5)


show(p)
```
